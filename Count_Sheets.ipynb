```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize = (10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('Image.jpg',0)
display(img)
```

```python
kernel=np.ones((151,1),dtype = np.float32) / (151*3)
im2 = cv2.filter2D(img,-1,kernel)
display(im2)
```

```python
im3 = cv2.adaptiveThreshold(im2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,1)
display(im3)

im = cv2.Sobel(im3, cv2.CV_64FC1, 1, 0, ksize = 5, scale = 5)
display(im)
```
